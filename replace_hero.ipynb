# Notebook pour modifier l'image principale et le titre d'une page HTML

Ce notebook charge `index.html`, remplace l'image du hero par `images/index/PREMIERE_IMPRESSION/IMAGE_PRINCIPALE.jpg`, et met à jour le titre principal sur une seule ligne.

## Étapes
1. Charger le fichier HTML.
2. Identifier la section hero et l'image existante.
3. Remplacer l'attribut `src` par le nouveau chemin.
4. Mettre à jour le titre principal.
5. Sauvegarder et vérifier.

```python
from pathlib import Path

path = Path('index.html')
text = path.read_text(encoding='utf-8')

# Remplacer le titre principal sur une seule ligne
old_title = '      <h1 class="hero-title">Transformez vos idées en<br>une véritable ambiance</h1>'
new_title = '      <h1 class="hero-title">Transformez vos idées en une véritable ambiance</h1>'
text = text.replace(old_title, new_title, 1)

# Remplacer l'image hero inline base64 par le fichier IMAGE_PRINCIPALE.jpg
import re
pattern = r'      <img src="data:image/jpeg;base64,[^"]*"'
text = re.sub(pattern, '      <img src="images/index/PREMIERE_IMPRESSION/IMAGE_PRINCIPALE.jpg" alt="Transformez vos idées en une véritable ambiance"', text, count=1)

path.write_text(text, encoding='utf-8')
print('Hero title and image updated in index.html')
```
